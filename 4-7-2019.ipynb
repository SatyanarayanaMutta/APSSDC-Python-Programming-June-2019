```python
# Graph to Plot distance and time

distance = list(range(2,11,2))
time=list(range(1,6))

plt.plot(time,distance)
plt.title("Distance VS Time")
plt.show()
```

```python
# Graph to Plot distance and time

distance = list(range(2,11,2))
time=list(range(1,6))

plt.plot(time,distance,'bo')
plt.grid()
plt.title("Distance VS Time")
plt.show()
```

```python
# Graph to Plot distance and time

distance = list(range(2,11,2))
time=list(range(1,6))

plt.plot(time,distance,"go")
plt.title("Distance VS Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid()
plt.show()
```

```python
# v = a t
t=list(range(1,11))
v=[2*i for  i in t]
#t=[2*i for i in t]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# v = a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for  i in range(len(t))]

plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# Numbers VS Cubes

n = list(range(10))
sqn = [i**3 for i in n]
plt.plot(n,sqn)

plt.show()
```

```python
# Voltage -Time

#t = [i/10 for i in range(1,21)]
import numpy as np
t=np.arange(0.1,2.0,0.1)
v= [1+np.sin(2*np.pi*i) for i in t]
plt.title("TIme VS Voltage Graph")
plt.grid()
plt.plot(t,v)
plt.xlabel("Time")
plt.ylabel("Voltaage")
plt.savefig("images/Voltage-Time.png")
plt.show()
            
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv("DataFiles/income.csv")
years = df.columns[2:]
#years = list(range(2005,2014))
incomeArkansas =df.values[3,2:]
plt.title("Income of Arkansas")
plt.xlabel("Years")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of All States from 2005 to 2013

years = df.columns[2:]

avg=df.values[0,2:]
#avg=[sum(list(map(int,i[2:])))/len(i[2:]) for i in df.values]# Not year wise Average
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.xlabel("Year")
plt.title("All States Income")
plt.ylabel("Income")
plt.plot(years,incomeallstates)
plt.grid()
plt.show()
```
